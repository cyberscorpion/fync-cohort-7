# Seaborn

```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python


# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="day",
    hue="smoker", style="smoker", size="size",
)
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.set_theme()
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
col1 = np.linspace(1,10, 1000)
col2 = np.sin(col1)
df1 = pd.DataFrame({"C1": col1, "C2": col2})
df1.head()
```

```python
sns.lineplot(x = df1.C1, y = df1.C2, data = df1)
plt.show()
```

```python
plt.figure(figsize = (5,3))
sns.set(rc = {'axes.facecolor' : 'cornflowerblue',
              'axes.grid' : False,
              'xtick.labelsize': 14,
              'ytick.labelsize' : 14
             })
sns.lineplot(x = df1.C1, y = df1.C2, data = df1)
plt.show()
```

## Automobile dataset

```python
df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(x = "wheel-base", y = "engine-size", data = df, linewidth=5)
plt.show()
```

```python
%matplotlib inline
```

```python

plt.figure(figsize=(10,5))
sns.lineplot(data = df["wheel-base"], linewidth=1.5, label='weel base')
sns.lineplot( data = df["engine-size"], linewidth=1.5, label='engine size')

plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(x = "wheel-base", y = "engine-size", data = df, linewidth=1.5, hue="fuel-type", style="fuel-type", markers=True)
plt.show()
```

## Fifa dataset

```python
df = pd.read_csv('fifa.csv', encoding = 'latin-1')
df.head()
```

```python
df.info()
```

```python
df['Body Type'].value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(5,4))
sns.distplot(df['Age'], bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(5,4))
x = df['Age']
ax = sns.distplot(x, bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(5,4))
x = df['Age']
ax = sns.distplot(x, bins=10, vertical = True)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(5,4))
x = df['Age']
ax = sns.kdeplot(x, shade=True, color = 'r')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(5,4))
x = df['Age']
ax = sns.distplot(x, kde=False , rug = True, bins=10)
plt.show()
```

```python
df['Preferred Foot'].nunique()
```

```python
df['Preferred Foot'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x = 'Preferred Foot', data = df, color = 'c')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x = 'Preferred Foot', hue="Real Face", data = df)
plt.show()
```

```python
df['International Reputation'].nunique()
```

```python
df['International Reputation'].value_counts()
```

```python
sns.stripplot(x = "International Reputation", y = "Potential", data = df, jitter = 0.01)
plt.show()
```

```python
sns.stripplot(x = "International Reputation", y = "Potential", data = df, hue="Preferred Foot", jitter = 0.2, palette='Set2', dodge=True)
plt.show()
```

```python
sns.stripplot(x = "International Reputation", y = "Potential", data = df,
              hue="Preferred Foot", jitter = 0.2, palette='Set2', size=20,
               marker = "D", alpha=.25, edgecolor = 'gray')
plt.show()
```

```python
sns.boxplot(df['Potential'])
plt.show()
```

```python
sns.boxplot(x = "International Reputation", y = "Potential", data = df,
              hue="Preferred Foot", palette = "Set3")
plt.show()
```

### Violin plot

### Point plot

```python
sns.pointplot(x = "International Reputation", y = "Potential", data = df, palette = "Set3")
plt.show()
```
